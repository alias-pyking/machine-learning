```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
#  Define the instruments to download
companies_dict = {
    'Amazon':'AMZN',
    'Apple':'AAPL',
    'Walgreen' : 'WBA',
    'Northrom Grumman':'NOC',
    'Boeing':'BA',
    'Lockheed Martin':'LMT',
    'McDonalds':'MCD',
    'Intel': 'INTC',
    'Navistar':'NAV',
    'IBM':'IBM',
    'Texas Instruments':'TXN',
    'MasterCard':'MA',
    'Microsoft':'MSFT',
    'General Electrics':'GE',
    'Symantec':'SYNC',
    'American Express':'AXP',
    'Pepsi':'PEP',
    'Coca Cola':'KO',
    'Johnson & Johnson':'JNJ',
    'Toyata':'TM',
    'Honda':'HMC',
    'Mistubishi':'MSBHY',
    'Sony':'SNE',
    'Exxon':'XOM',
    'Chevron':'CVX',
    'Valero Energy':'VLO',
    'Ford':'F',
    'Bank of America':'BAC'}

companies = sorted(companies_dict.items(), key=lambda x:x[1])

print(companies)
```

```python
# Define which online source to use
source = "yahoo"
# define the start and end dates
start_date = '2015-01-01'
end_date = '2017-12-31'
# use pandas_reader to load the desire stock data
values = [x for x in companies_dict.values()]
print(values)
panel_data = data.DataReader(values,data_source=source,start= start_date,end =end_date)

# Print Axes labels
```

```python
stock_close = panel_data[['Close']]
stock_open = panel_data[['Open']]
# print(stock_close.iloc[0])
stock_close
```

```python
# Calculate daily stock movement
stock_close = np.array(stock_close).T
stock_open = np.array(stock_open).T
row,col = stock_close.shape
print(row, col)
movements = np.zeros([row,col])
for i in range(0,row):
    movements[i,:] = np.subtract(stock_close[i,:],stock_open[i,:])
```

```python

for i in range(len(companies)):
    print('Company: {},  movement: {} '.format(companies[i][0],sum(movements[i,:])))
```

```python
# Visualizing the movements of companies
plt.clf
plt.figure(figsize=(20,20))
ax1 = plt.subplot(221)
plt.plot(movements[0,:])
plt.title(companies[0])
plt.subplot(222, sharey = ax1)
plt.plot(movements[1][:])
plt.title(companies[1])
plt.show()
```

```python
# Import Normalizer
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()
new = normalizer.fit_transform(movements)
print(new.max())
print(new.mean())
print(new.min())
```

```python
plt.clf
plt.figure(figsize=(20,20))
ax1 = plt.subplot(221)
plt.plot(new[0,:])
plt.title(companies[0])
plt.subplot(222, sharey = ax1)
plt.plot(new[1][:])
plt.title(companies[1])
plt.show()
```

```python
# Import necessary libraries
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer
# Define the normalizer 
normalizer = Normalizer()
# create a KMeans model with 10 cluster

k_means = KMeans(n_clusters=10, max_iter=1000)

# Make a pipeline chaining normalizer and kmeans
pipeline  = make_pipeline(normalizer, k_means)
```

```python
# Fit pipeline to daily stock movements
pipeline.fit(movements)
print(k_means.inertia_)
```

```python
# Import pandas 
import pandas as pd
# Predict the cluster label
labels = pipeline.predict(movements)

# Create a DataFrame aligning companies and labels
df = pd.DataFrame({'labels':labels,'companies':companies})

# Display values sorted by cluster labels
print(df.sort_values('labels'))

print(labels)
print(len(labels))
```

```python
# Do a dimensionality reduction on the higher dimensional data to do visualization of clusters
from sklearn.decomposition import PCA
reduced_data = PCA(n_components=2).fit_transform(new)

# run k_means on the reduced data
kmeans = KMeans(n_clusters=10)
kmeans.fit(reduced_data)
labels = kmeans.predict(reduced_data)
# Create a DataFrame aligning labels and companies
df = pd.DataFrame({'labels':labels,'companies':companies})
print(df.sort_values('labels'))
```

```python
# Define the step size of mesh
h = 0.01
x_min, x_max = reduced_data[:,0].min() - 1,reduced_data[:,0].max() + 1
y_min, y_max = reduced_data[:,1].min() - 1, reduced_data[:,1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min,x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point  in mesh using our trained model
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
# Put the results into a colour plot
Z = Z.reshape(xx.shape)
# Define colour plot 
cmap = plt.cm.Paired
plt.clf()
plt.figure(figsize = (10,10))
plt.imshow(Z, interpolation='nearest', extent = (xx.min(), xx.max(), yy.min(), yy.max()),
          cmap=cmap,aspect='auto', origin = 'lower')
plt.plot(reduced_data[:,0], reduced_data[:,1],'k.', markersize=5)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='x',s = 169, linewidths=3, color = 'w', zorder= 10)
plt.title('Kmeans clustering on Stock Market Movements (PCA Reduces Data) ')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
