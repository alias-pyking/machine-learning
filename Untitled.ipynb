```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
companies_dict = {
    'Amazon':'AMZN',
    'Apple':'AAPL',
    'Walgreen' : 'WBA',
    'Northrom Grumman':'NOC',
    'Boeing':'BA',
    'Lockheed Martin':'LMT',
    'McDonalds':'MCD',
    'Intel': 'INTC',
    'Navistar':'NAV',
    'IBM':'IBM',
    'Texas Instruments':'TXN',
    'MasterCard':'MA',
    'Microsoft':'MSFT',
    'General Electrics':'GE',
    'Symantec':'SYNC',
    'American Express':'AXP',
    'Pepsi':'PEP',
    'Coca Cola':'KO',
    'Johnson & Johnson':'JNJ',
    'Toyata':'TM',
    'Honda':'HMC',
    'Mistubishi':'MSBHY',
    'Sony':'SNE',
    'Exxon':'XOM',
    'Chevron':'CVX',
    'Valero Energy':'VLO',
    'Ford':'F',
    'Bank of America':'BAC'}
```

```python
companies  = sorted(companies_dict.items(), key = lambda x:x[1])
companies
```

```python
source = 'yahoo'
start_date = '2015-01-01'
end_date = '2017-12-31'
values = [x for x in companies_dict.values()]
data_set = data.DataReader(values,data_source=source, start=start_date,end=end_date)
```

```python
data_set.describe()
```

```python
data_set.shape
```

```python
data_set
```

```python
stock_open = data_set[['Open']]
stock_close = data_set[['Close']]
stock_close.fillna(0.0)
stock_open.fillna(0.0)
```

```python
stock_open = np.array(stock_open).T
stock_close = np.array(stock_close).T

row, col = stock_close.shape
print(row,col)
```

```python
movements = np.zeros([row,col])
for i in range(row):
    movements[i,:] = np.subtract(stock_close[i,:],stock_open[i,:])
movements.shape
```

```python
for i in range(len(companies)):
    print('Company : {} , movement : {} '.format(companies[i][0],sum(movements[i,:])))
```

```python
plt.clf()
plt.figure(figsize=(20,20))
ax1 = plt.subplot(221)
plt.plot(movements[5,:])
plt.title(companies[5])
plt.subplot(222, sharey = ax1)
plt.plot(movements[1,:])
plt.title(companies[1])
plt.show()
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
new = normalizer.fit_transform(movements)
print(new.max())
print(new.mean())
print(new.min())
```

```python
plt.clf()
plt.figure(figsize=(20,20))
ax1 = plt.subplot(221)
plt.plot(new[0,:])
plt.title(companies[0])
plt.subplot(222, sharey = ax1)
plt.plot(new[1,:])
plt.title(companies[1])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
kmeans.fit(new)
labels = kmeans.predict(new)
df = pd.DataFrame({'labels':labels,'companies':companies})
print(df.sort_values('labels'))
print(kmeans.inertia_)
```
